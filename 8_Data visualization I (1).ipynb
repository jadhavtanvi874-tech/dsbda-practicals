```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv('Titanic-Dataset.csv')

# Calculate skewness for the 'Fare' column
# Note: Use capital 'F' as it appears in your CSV
fare_skewness = dataset['Fare'].skew()

print(f"The skewness of the Fare column is: {fare_skewness}")

# Logic to interpret the result
if fare_skewness > 0:
    print("The distribution is Positively Skewed (Right-skewed).")
elif fare_skewness < 0:
    print("The distribution is Negatively Skewed (Left-skewed).")
else:
    print("The distribution is perfectly symmetrical.")
```
