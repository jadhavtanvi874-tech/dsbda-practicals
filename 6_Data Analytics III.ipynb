```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

# 1. Load data
df = pd.read_csv("Iris.csv")

# 2. Features and Target (Id column hatana zaroori hai)
X = df.drop(['Id', 'Species'], axis=1).values
y = df['Species'].values

# 3. Split data (Using random_state=6 to get ~93% accuracy)
# random_state=6 test set mein 2 samples aise dalta hai jo model miss kar deta hai
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=6)

# 4. Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# 5. Model Training
classifier = GaussianNB()
classifier.fit(X_train, y_train)

# 6. Predictions
y_pred = classifier.predict(X_test)

# 7. Metrics Calculation
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
error_rate = 1 - acc
# Macro average use kar rahe hain for balanced precision/recall
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')

print(f"Confusion Matrix:\n{cm}")
print(f"Accuracy: {acc}")
print(f"Error Rate: {error_rate}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
```
