```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# 1. Data Load karna
df = pd.read_csv('Social_Network_Ads (1).csv')

# 2. Features (X) aur Target (y) select karna
X = df[['Age', 'EstimatedSalary']]
y = df['Purchased']

# 3. Data Split (Training aur Testing mein)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# 4. Classification Model banana (Logistic Regression)
model = LogisticRegression()
model.fit(X_train, y_train)

# 5. Prediction karna
y_pred = model.predict(X_test)

# 6. Confusion Matrix nikalna
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()

# 7. Metrics Calculate karna (Jo question mein pucha hai)
accuracy = accuracy_score(y_test, y_pred)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
error_rate = (fp + fn) / (tp + tn + fp + fn)

# Results Print karna
print("Confusion Matrix:\n", cm)
print(f"Accuracy: {accuracy:.2f}")
print(f"Error Rate: {error_rate:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```

```python
#
Accuracy
Definition: It is the ratio of the total number of correct predictions to the total number of input samples. It shows how often the model is correct overall.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$

Precision
Definition: It is the ratio of correctly predicted positive observations to the total predicted positive observations. It answers: "Of all people predicted to buy, how many actually bought?"
Formula:
$$\text{Precision} = \frac{TP}{TP + FP}$$

Recall (Sensitivity)
Definition: It is the ratio of correctly predicted positive observations to all observations in the actual class. It answers: "Of all people who actually bought, how many did the model identify?"
Formula:
$$\text{Recall} = \frac{TP}{TP + FN}$$

Error Rate
Definition: It is the ratio of the total number of incorrect predictions to the total number of samples. It is the exact opposite of accuracy.
$$\text{Error Rate} = \frac{FP + FN}{TP + TN + FP + FN}$$

```
