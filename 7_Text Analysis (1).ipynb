```python
import math
import pandas as pd

# 1. Sabse pehle word counts (BoW) prepare karein
# (Aapki file ke Cell 15 ka fixed version)
wordDictA = dict.fromkeys(total, 0)
wordDictB = dict.fromkeys(total, 0)

for word in first_sentence:
    wordDictA[word] += 1
for word in second_sentence:
    wordDictB[word] += 1 # Bhai yahan pehle sirf 'wordDictB' likha tha, ab increment ho raha hai

# 2. TF (Term Frequency) Function
def computeTF(wordDict, doc):
    tfDict = {}
    corpusCount = len(doc)
    for word, count in wordDict.items():
        tfDict[word] = count / float(corpusCount)
    return tfDict

tfFirst = computeTF(wordDictA, first_sentence)
tfSecond = computeTF(wordDictB, second_sentence)

# 3. IDF (Inverse Document Frequency) Function - FIXED
def computeIDF(docList):
    idfDict = {}
    N = len(docList)
    
    # Pehle check karo ki word kitne docs mein hai
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    # Ab log formula lagao
    for word, val in idfDict.items():
        # N / val: Total docs / Docs containing the word
        idfDict[word] = math.log10(N / float(val)) 
    return idfDict

idfs = computeIDF([wordDictA, wordDictB])

# 4. TF-IDF Function
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val * idfs[word]
    return tfidf

tfidfFirst = computeTFIDF(tfFirst, idfs)
tfidfSecond = computeTFIDF(tfSecond, idfs)

# 5. Output dekhne ke liye DataFrame
df = pd.DataFrame([tfidfFirst, tfidfSecond])
print("--- Final TF-IDF Output ---")
print(df)
```
