```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
text = "NLTK is great, but these errors are a bit annoying!"
tokenized_sent = word_tokenize(text.lower())
stop_words = set(stopwords.words("english"))
filtered_sent = []
for w in tokenized_sent:
    if w not in stop_words:
        filtered_sent.append(w)
print("Tokenized:", tokenized_sent)
print("Filtered:", filtered_sent)
```

```python
import pandas as pd
first_sentence = "Data Science is the sexiest job of the 21st century"
second_sentence = "machine learning is the key for data science"
wordsetA = first_sentence.split(" ")
wordsetB = second_sentence.split(" ")
total_vocab = set(wordsetA).union(set(wordsetB))
wordDictA = dict.fromkeys(total_vocab, 0)
wordDictB = dict.fromkeys(total_vocab, 0)
for word in wordsetA:
    wordDictA[word] += 1
for word in wordsetB:
    wordDictB[word] += 1
df = pd.DataFrame([wordDictA, wordDictB])
print(df)
```

```python
def computeTF(wordDict, doc):
    tfDict = {}
    corpusCount = len(doc)
    for word, count in wordDict.items():
        tfDict[word] = count / float(corpusCount)
    return tfDict
tfFirst = computeTF(wordDictA, wordsetA)
tfSecond = computeTF(wordDictB, wordsetB)
import pandas as pd
tf_df = pd.DataFrame([tfFirst, tfSecond])
print(tf_df)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filtered_sentence = [w for w in wordDictA if w.lower() not in stop_words]
print("Filtered Words:", filtered_sentence)
```

```python
import math
def computeIDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    all_words = set().union(*[doc.keys() for doc in docList])
    idfDict = dict.fromkeys(all_words, 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
    return idfDict
idfs = computeIDF([wordDictA, wordDictB])
print("IDF Values:", idfs)
```

```python
import pandas as pd
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val * idfs[word]
    return tfidf
tfidfFirst = computeTFIDF(tfFirst, idfs)
tfidfSecond = computeTFIDF(tfSecond, idfs)
tfidf_df = pd.DataFrame([tfidfFirst, tfidfSecond])
print(tfidf_df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
first_sentence = "Data Science is the sexiest job of the 21st century"
second_sentence = "machine learning is the key for data science"
vectorizer = TfidfVectorizer()
response = vectorizer.fit_transform([first_sentence, second_sentence])
feature_names = vectorizer.get_feature_names_out()
df_results = pd.DataFrame(response.toarray(), columns=feature_names)
print(df_results)
```
