```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("iris.csv")
df.columns = ['col1', 'col2', 'col3', 'col4', 'col5', 'col6']
df.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))

axes[0,0].set_title('Distribution of First Column')
axes[0,0].hist(df['col2']);

axes[0,1].set_title('Distribution of Second Column')
axes[0,1].hist(df['col3']);

axes[1,0].set_title('Distribution of Third Column')
axes[1,0].hist(df['col4']);

axes[1,1].set_title('Distribution of Fourth Column')
axes[1,1].hist(df['col5']);

plt.tight_layout();
plt.show()
```

```python
df.iloc[:, 0:5] = df.iloc[:,0:5].apply(pd.to_numeric, errors='coerce')
data_to_plot = [df['col2'], df['col3'], df['col4'], df['col5']]
sns.set_style("whitegrid")
fig = plt.figure(1, figsize=(12, 8))
ax = fig.add_subplot(111)
bp = ax.boxplot(data_to_plot)
plt.show()
```
