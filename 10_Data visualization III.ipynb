```python
plt.figure(figsize=(6, 6))
sns.histplot(df['SepalLengthCm'],color="skyblue",kde="true") 
plt.title('Simple Histogram')
plt.show()
```

```python
# Boxplot banayein
plt.figure(figsize=(6, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=df,color="skyblue")
plt.title('Simple Boxplot by Species')
plt.show()
```

```python
print("--- Data Distribution ---")
print(df.describe()) # Isse Mean, Min, Max aur Quartiles mil jayenge

# 3. Outlier Detect karna (Sirf SepalWidth ke liye, kyunki isi mein outliers hain)
Q1 = df['SepalWidthCm'].quantile(0.25)
Q3 = df['SepalWidthCm'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Outliers ko filter karke dekhna
outliers = df[(df['SepalWidthCm'] < lower) | (df['SepalWidthCm'] > upper)]
print("\nDetect huye Outliers:")
print(outliers[['SepalWidthCm']])

# 4. Outliers ko visualize karna (Boxplot)
sns.boxplot(x=df['SepalWidthCm'], color="cyan")
plt.title('Outliers in Sepal Width')
plt.show()
```
